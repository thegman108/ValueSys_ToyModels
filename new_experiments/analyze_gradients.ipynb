```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
#pickle_in = open("../../gradients_sparse_epoch_0.pkl","rb")

def load_gradients(training_type):
    with open('../gradients/gradients_sparse_epoch_0.pkl', 'rb') as f:
        gradients = pickle.load(f)
    return gradients

# Load gradients
gradients_sparse = load_gradients('sparse')  # or 'dense' based on the training type

# Analyze and plot gradients

# Extract valid gradients from the loaded data, filtering out NaN values
valid_gradients_sparse = [g for g in gradients_sparse if not np.isnan(g).any()]

```

```python


# Convert the valid gradients list to a numpy array for easier manipulation
gradients_array_0 = np.array(valid_gradients_sparse)

# Create a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(gradients_array_0, cmap='viridis', cbar=True)
plt.title('Heatmap of Gradient Norms Over Training Steps Sparse 0')
plt.xlabel('LoRA Parameter Index')
plt.ylabel('Training Steps')
plt.show()
```

```python
def plot_gradients(gradients, title):
    avg_gradients = [np.mean(g) for g in gradients]
    plt.plot(avg_gradients)
    plt.title(title)
    plt.xlabel('Training Steps')
    plt.ylabel('Average Gradient Magnitude')
    plt.show()

plot_gradients(valid_gradients_sparse, 'Sparse Gradients Over Time')  # or 'Dense Gradients Over Time' based on the training type
```
